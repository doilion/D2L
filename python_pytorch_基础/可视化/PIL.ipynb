### 2、读取网络图片

```python
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt
import requests

url = "https://farm4.staticflickr.com/3120/2862048068_0cf5eed770_z.jpg"
image = Image.open(requests.get(url, stream=True).raw)

print(image.size)

plt.imshow(image)
plt.axis("off")
plt.show()
```
