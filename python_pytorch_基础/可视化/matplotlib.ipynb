### 1、折线图

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.5)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='y1', color='r', marker='*', linestyle='-')
plt.plot(x, y2, label='y2', color='b', marker='o', linestyle='--')
plt.legend(loc='upper left')
plt.show()
```

### 2、散点图

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制散点图
plt.scatter(x, y)

# 添加标题和坐标轴标签
plt.title("scatter plot")
plt.xlabel("x")
plt.ylabel("y")

# 显示网格线
plt.grid()

plt.show()
```

### 3、绘制矩形框

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 创建一个举行的对象
rectangle = patches.Rectangle((0.2, 0.3), 0.5, 0.4, edgecolor='r', facecolor='none')

# 显示图形
fig, ax = plt.subplots()
ax.add_patch(rectangle)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

### 3、图像读取与可视化

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 读取图像
image = mpimg.imread('example_image.jpg')

# 获取图像尺寸  H W C
height, width, channels = image.shape
print(height, width, channels)

# 可视化图像
plt.imshow(image)
plt.axis('off') # 关闭坐标轴
plt.show()
```

```python
# 同时这里也可以使用 PIL 对图像进行一个可视化
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt

image = Image.open('example_image.jpg')

plt.axis('off')
plt.imshow(image)
plt.show()
```

### 4、可视化多张图像

```python
%matplotlib inline
import PIL.Image as Image
import matplotlib.pyplot as plt

img1 = Image.open('2024-07-01-033305.jpg')
img2 = Image.open('2024-07-01-033308.jpg')
img3 = Image.open('2024-07-01-033312.jpg')

plt.subplot(221)
plt.axis('off')
plt.imshow(img1)

plt.subplot(222)
plt.axis('off')
plt.imshow(img2)

plt.subplot(223)
plt.axis('off')
plt.imshow(img3)

plt.show()
```

###  5、在图像上绘制散点

```python
%matplotlib inline
import matplotlib.pyplot as plt
import PIL.Image as Image
import numpy as np

# 读取图像
image = Image.open('example_image.jpg')
width, height = image.size[:2]

# 绘制图像
plt.imshow(image)

# 生成一些随机的点
num_points = 50
x = np.random.randint(0, width,  num_points)
y = np.random.randint(0, height, num_points)

# 绘制散点图
plt.scatter(x, y, color='r', marker='o')

# 显示图像
plt.axis('off')
plt.show()
```

### 6、在图像上绘制矩形 并标注文字

```python
%matplotlib inline
from PIL import Image
import requests
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 读取一张图像
url = 'https://farm4.staticflickr.com/3120/2862048068_0cf5eed770_z.jpg'
image = Image.open(requests.get(url, stream=True).raw)

# 绘制图像
plt.axis('off')
plt.imshow(image)

# 创建一个矩形对象并添加到图像中，(66, 110) 表示的是左上角的 x 和 y
rectangle = patches.Rectangle((66, 110), 85, 113, linewidth=2, edgecolor='r', facecolor='none')
plt.gca().add_patch(rectangle)

# 标注文字
plt.text(66, 110, 'cow: 86%', color='r', fontsize=9, verticalalignment='bottom',
         bbox=dict(facecolor='yellow', edgecolor='r'))

# 显示图像
plt.show()
```
